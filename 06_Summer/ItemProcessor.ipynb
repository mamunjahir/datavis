```python
def item_process1(data2):
    dict1={}
    arr_columns=[]
    arr=[]
    words=['student','school']
    for columns in data2.columns:
        #print(columns)
        #arr_columns.append(columns)
        #init_row=''
        #print(columns)
        for r in data2[columns]:
            if isinstance(r, str):
                rows=r.lower()
                for i in words:
                    if i in rows:
                        if rows not in dict1:
                            dict1[rows]=1
                            #print (rows)
                        else:
                            dict1[rows]=dict1[rows]+1
                        #print(rows)
                        arr.append(rows)
    for key in dict1:
        #print (key)
        if dict1[key]>1:
            print(key,dict1[key])
    return dict1;
```
